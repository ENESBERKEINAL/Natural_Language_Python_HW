```
import nltk
from nltk import ngrams
from nltk.tokenize import word_tokenize
from nltk.tokenize import RegexpTokenizer
nltk.download('averaged_perceptron_tagger')
from collections import Counter
```

```
data="Tina was a 6 years old cute girl. She is very fond of wooden toys, especially she had a beautiful wood horse gifted by her uncle when she was 2 years old. The wooden horse has been her close pal and her pet. She has a 9 years old brother. She went on a vacation to a nature resort in the woods with her family. She carried the wooden horse with her. She enjoyed her holiday with her family in the woods. While she was packing things with her brother as they planned to return home, the wooden horse fell down and one of the legs was broken. Tina was too sad and silently cried for her horse. She was very upset. They packed all things and left the woods. The entire family tried to cheer up the sweet little girl, but Tina was so silent, very upset. Tina’s brother tried to console her a lot. They had a break for lunch and Tina refused to eat. Her momma requested her to eat food, she ate a very little amount of food. While others were eating, she sat quietly in their car. Her brother came to her and kissed her cheek, told her, “Tina dear, don’t worry, don’t get upset dear. It is only a wooden toy. The horse doesn’t have a life like us and it is just a lifeless thing. Don’t be so sad about the broken leg. Even if the horse loses its tail, it won’t be hurting the horse. Even if all four legs are broken, the wooden horse remains the same. Even if the horse loses its head, it won’t be painful. I will buy a new wood horse for you!” Tina replied, ‘You think it doesn’t matter that a leg of my pet toy was broken?’ Her brother replied, ‘Yes dear!’ Tina replied, ‘Yes brother you are right. It won’t be a major issue for you even if the horse is broken into pieces at this age. But if you are in my age, if you had a pet like I had, you will feel how painful it would be to even if a small part of the horse is broken!’ Her brother left silently!"
data=data.lower()
data
```

```
"""
def get_ngrams(text, n ):
    n_grams = ngrams(word_tokenize(text), n)
    return [ ' '.join(grams) for grams in n_grams]"""
  
def get_ngrams(text, n ):
    tokenizer = RegexpTokenizer(r'\w+')
    tokens=tokenizer.tokenize(text)
    n_grams = ngrams(tokens, n)
    return [ ' '.join(grams) for grams in n_grams]
```

```
def WordList(data):
  word_list = [];
  for i in range(len(data)):
    count = 0;
    isThereOnList = False
    for j in range(len(data)):
      if data[i] == data[j]:
        count = count + 1
    for k in range(len(word_list)):
      if word_list[k][0] == data[i]:
        isThereOnList = True
    if isThereOnList != True:
      word_list.append([data[i], count])
  return word_list
```

```
def MoreThanOne(data):
  more_than_one = []
  for i in range(len(data)):
    if data[i][1] != 1:
      more_than_one.append(data[i])
  return more_than_one
```

```
one_grams = get_ngrams(data, 1)
one_grams
```

```
one_gram_list = WordList(one_grams)
one_gram_list
```

```
two_grams = get_ngrams(data, 2)
two_grams
```

```
two_gram_list = WordList(two_grams)
two_gram_list
```

```
MoreThanOne(two_gram_list)
```

```
tri_grams = get_ngrams(data, 3)
tri_grams
```

```
tri_gram_list = WordList(tri_grams)
tri_gram_list
```

```
MoreThanOne(tri_gram_list)
```
